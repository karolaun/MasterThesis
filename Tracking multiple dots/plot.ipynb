```python
file_path = "./ResultsTwice/FiltRes.txt"  # Replace with your file path
coordinates = []
with open(file_path, "r") as file:
    for line in file:
        line = line.strip()[1:-1]  # Remove brackets and strip whitespace
        x, y = map(float, line.split())  # Split the line and convert values to float
        coordinates.append((x, y))

# Separate x and y coordinates into lists
x_values = [coord[0] for coord in coordinates]
y_values = [coord[1] for coord in coordinates]

# Plot the coordinates
plt.scatter(x_values, y_values, color='blue')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('Scatter Plot of Coordinates')
```

```python
import matplotlib.pyplot as plt
from collections import defaultdict

# Define a function to parse the data and separate into groups based on index
def parse_data(file_path):
    data_dict = defaultdict(list)
    with open(file_path, 'r') as file:
        for line in file:
            parts = line.strip().split(' : ')
            if len(parts) == 2:
                index = int(parts[0])
                coordinates = [int(coord) for coord in parts[1].split(' , ')]
                data_dict[index].append(coordinates)
    return data_dict

# Function to plot each group separately
def plot_groups(data_dict):
    for index, points_list in data_dict.items():
        if len(points_list) > 0:
            points_list = list(zip(*points_list))
            plt.scatter(points_list[0], points_list[1], label=f'Index {index}')
    plt.xlabel('X-coordinate')
    plt.ylabel('Y-coordinate')
    plt.title('Scatter Plot of Groups')
    plt.legend()
    plt.show()

# Example usage
file_path = "./Week13/FilteringChatGpt/resred.txt"
data_dict = parse_data(file_path)
plot_groups(data_dict)

for index, values in data_dict.items():
    x = [val[0] for val in values]
    y = [val[1] for val in values]

    plt.figure()  # Create a new figure for each index
    plt.scatter(x, y)
    plt.title(f"Index {index} Scatter Plot")
    plt.xlabel("X values")
    plt.ylabel("Y values")
    plt.grid(True)
    plt.show()
```
