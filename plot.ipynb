```python
# Initialize lists to store x and y values
x_values = []
y_values = []

# Read data from file
with open('./Week14/Green/res0.txt', 'r') as file:
    for line in file:
        values = line.strip().split(',')
        if len(values) == 2:
            try:
                x_values.append(float(values[0].strip()))
                y_values.append(float(values[1].strip()))
            except ValueError:
                print("Warning: Cannot convert to float")

# Calculate quartiles for x and y separately
x_Q1 = np.percentile(x_values, 0.1)
x_Q3 = np.percentile(x_values, 99.9)

y_Q1 = np.percentile(y_values, 0.1)
y_Q3 = np.percentile(y_values, 99.9)

# Calculate interquartile range (IQR) for x and y separately
x_IQR = x_Q3 - x_Q1
y_IQR = y_Q3 - y_Q1

# Define outlier range for x and y separately (e.g., 1.5 times IQR)
k = 1.5
x_lower_limit = x_Q1 - k * x_IQR
x_upper_limit = x_Q3 + k * x_IQR

y_lower_limit = y_Q1 - k * y_IQR
y_upper_limit = y_Q3 + k * y_IQR

# Remove values outside the limits in both x and y directions
filtered_x_values = []
filtered_y_values = []

for x, y in zip(x_values, y_values):
    if x_lower_limit <= x <= x_upper_limit and y_lower_limit <= y <= y_upper_limit:
        filtered_x_values.append(x)
        filtered_y_values.append(y)

print("Original x values:", x_values)
print("Filtered x values:", filtered_x_values)
print("Original y values:", y_values)
print("Filtered y values:", filtered_y_values)


# Plot the coordinates
plt.scatter(filtered_x_values, filtered_y_values, color='blue')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
#plt.ylim((-535,-525))
#plt.xlim((130,140))
plt.title('Scatter Plot of Coordinates')
plt.savefig("ScatterSpreadres0Green.svg",format="svg")
plt.show()

# Create a 2D histogram
heatmap, xedges, yedges = np.histogram2d(x_values, y_values, bins=100)

# Plot heatmap
plt.imshow(heatmap.T, extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]], origin='lower', cmap='inferno')
plt.colorbar(label='Density')
#plt.xlim((156.5,158.5))
#plt.ylim((836.5,839.5))
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Density Plot')
plt.savefig("Densityres0green.svg",format="svg")
plt.show()
```
